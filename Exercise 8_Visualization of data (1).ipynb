### Step 5. Create a plot to check the relationship between horsepower and acceleration. Note down your insight for the same. Beautify the graph using various customizations.

```python
mpg_df["horsepower"].value_counts()
```

```python
sns.catplot(x="horsepower",y="acceleration",data=mpg_df)
plt.xlabel("Horsepower", fontsize=15, color="Red")
plt.ylabel("Acceleration", fontsize=15, color="Red")
plt.title("Relationship bet. horsepower and acceleration", fontsize=20, color="Brown")
plt.show()
```

```python
## Swarm Plot
sns.catplot(x="horsepower",y="acceleration",kind="swarm",data=mpg_df)
plt.xlabel("Horsepower", fontsize=15, color="Red")
plt.ylabel("Acceleration", fontsize=15, color="Red")
plt.title("Relationship bet. horsepower and acceleration", fontsize=20, color="Brown")
plt.show()
```

```python
## Insight: Majority of cars have less Horspower and Acceleration from the mean. Othwewise data looks symmetrical
## and there is no particular relation between two attributes
```

```python
## Viloin Plot
sns.catplot(x="horsepower",y="acceleration",kind="violin",data=mpg_df)
plt.xlabel("Horsepower", fontsize=15, color="Red")
plt.ylabel("Acceleration", fontsize=15, color="Red")
plt.title("Relationship bet. horsepower and acceleration", fontsize=20, color="Brown")
plt.show()
```

```python
## Box Plot
sns.catplot(x="horsepower",y="acceleration",kind="box",data=mpg_df)
plt.xlabel("Horsepower", fontsize=15, color="Red")
plt.ylabel("Acceleration", fontsize=15, color="Red")
plt.title("Relationship bet. horsepower and acceleration", fontsize=20, color="Brown")
plt.show()
```

### Step 9. Check the relationship of multiple variables wrt kilometer_per_litre. Note down your insight for the same

```python
kilometer_per_litre = 0.4251* mpg_df["mpg"]
```

```python
mpg_df["kilometer_per_litre"]=kilometer_per_litre
```

```python
mpg_df.dtypes
```

```python
sns.relplot(x="cylinders",y="kilometer_per_litre", data=mpg_df, palette="Set2")
plt.title("Relationship between number of cylinders and KPL")
plt.show()
```

```python
## Insight: As the number of cylinders increases, KPL decreases
```

```python
sns.relplot(x="displacement",y="kilometer_per_litre", data=mpg_df, palette="Set2")
plt.title("Relationship between displacement and KPL")
plt.show()
## Insight: As the displacement increases, KPL decreases
```

```python
sns.relplot(x="horsepower",y="kilometer_per_litre", data=mpg_df, palette="Set2")
plt.title("Relationship between Horsepower and KPL")
plt.show()
## Insight: There is no particular relationship between horsepower and KPL
```

```python
sns.relplot(x="weight",y="kilometer_per_litre", data=mpg_df, palette="Set2")
plt.title("Relationship between Weight and KPL")
plt.show()
## Insight: As the weight increases, KPL decreases
```

```python
sns.relplot(x="acceleration",y="kilometer_per_litre", data=mpg_df, palette="Set2")
plt.title("Relationship between acceleration and KPL")
plt.show()
## Insight: As the acceleration increases, KPL increases. But there are many outliers as well.
```

```python
sns.catplot(x="origin",y="kilometer_per_litre", data=mpg_df, palette="Set2", kind="swarm")
plt.title("Relationship between Origin and KPL")
plt.show()
## Insight: As the Origin increases, average KPL increases.
```

```python
sns.catplot(x="origin",y="kilometer_per_litre", data=mpg_df, palette="Set2", kind="violin")
plt.title("Relationship between Origin and KPL")
plt.show()
```

```python
sns.catplot(x="origin",y="kilometer_per_litre", data=mpg_df, palette="Set2", kind="box")
plt.title("Relationship between Origin and KPL")
plt.show()
```

### Step 10. Display the average weight based on no of cylinders present. Note down the insight for the same

```python
sns.catplot(x="cylinders",y="weight", data=mpg_df, kind="violin")
plt.title("Average Weight based on Number of Cylinders", color="red")
plt.show()
## Insight: As the number of cylinder increases, average weight increases
```
