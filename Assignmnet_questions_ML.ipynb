### What is the distribution  of preferred foot?


```python
sns.countplot(x="Preferred Foot", data=df)
plt.show()
## Majority of the players has right foot as the preferred foot
```

### How can we visualize the age distribution of players?

```python
sns.distplot(df["Age"])
plt.show()
## Distribution of age is close to normal distribution. Mean and median are very 
## close to each other. Mean is 25.122206 and median is 25
```

```python
df["Age"].describe()
```

```python
df["Age"].skew()
## There is very little skewness in the age column
```

```python
sns.boxplot(x="Age", data=df)
plt.show()
## There exists some outliers in the Age column as well
```

### What is the international reputation of left and right foot?

```python
sns.countplot(x="Preferred Foot", hue="International Reputation", data=df)
plt.show()
```

### What is the different work rate of the Players?

```python
df["Work Rate"].value_counts().plot(kind="bar")
plt.show()
## Most of the footballers have Medium/ Medium work rate followed by 
## High/ Medium and Medium/ High
```

### Top 10 countries with most number of players

```python
df["Nationality"].value_counts()[0:10]
```

### What is the work rate of players segregated by preferred foot?

```python
## Both work rate and preferred foot are categorical variables. So we're using Countplot here 
sns.countplot(x="Preferred Foot", hue="Work Rate", data=df)
plt.show()
```

### Number of players at different positions? 



```python
df["Position"].value_counts().plot(kind="bar")
plt.show()
```

### Age distribution in top countries

```python
countries=list(df["Nationality"].value_counts()[0:9].index)
```

```python
sns.distplot(df.loc[df["Nationality"].isin(countries),"Age"])
plt.show()
## Shows Age distribution in top 10 countries
```

```python
df.loc[df["Nationality"].isin(countries),"Age"].describe()
```

```python
df.loc[df["Nationality"].isin(countries),"Age"].skew()
```

### What is the distribution of overall score of players from top countries?

```python
sns.distplot(df.loc[df["Nationality"].isin(countries),"Overall"])
plt.show()
## Shows Overall score distribution in top 10 countries
## Looks like it is normally distributed
```

```python
df.loc[df["Nationality"].isin(countries),"Overall"].describe()
## Mean is almost equal to median. So it is normall distributed
```

```python
df.loc[df["Nationality"].isin(countries),"Overall"].skew()
```
