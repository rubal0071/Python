```python
## Whenever we're using cmap, we can use codes in colors because every number/ code represents some color
x=np.array([2,5,1,8,10,12])
y=np.array([45,67,80,40,90,100])
colors=np.array([10,20,30,40,50,100])
size=[500,600,700,800,900,1000]
plt.scatter(x,y,c=colors,s=size,cmap="viridis")
plt.show()
```
